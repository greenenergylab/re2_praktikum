## Diagramme zeichnen

Für das Praktikum, aber auch für viele andere Anwendungen als Ingenieur ist es wichtig, Messdaten nicht nur erfassen und auswerten zu können, sondern diese auch ansprechend Darstellen zu können. Diagramme sind eye-catcher und werden in Präsentationen, Berichten, Abschlussarbeiten und vielen weiteren Fällen verwendet. Wie dies mit Python funktioniert, wird nun näher erläutert.

Es gibt viele verschiedene Arten von Diagrammen: Säulendiagramme, Tortendiagramme, Histogramme, ...

Die verschiedenen Diagrammarten werden bei verschiedenen Arten von Daten verwendet. Tortendiagramme werden beispielsweise häufig verwendet, um die Grundgesamtheit darzustellen, also z.B. die elementare Zusammensetzung von Stahl.

Im Falle dieses Praktikums werden überwiegend Diagramme benötigt, die den Zusammenhang zwischen zwei veränderlichen Größen zeigen - sogenannte xy-Diagramme. Dabei gilt in der Wissenschaft, dass gemessene Daten**punkte** stets als Punkte dargestellt werden und nicht als verbindende Linie oder ähnlichem. 

Ferner gilt für xy-Diagramme, dass die *unabhängige Variable auf der x-Achse* und die *abhängige Variable auf der y-Achse* dargestellt werden. 

In unserem Beispiel aus dem letzten Kapitel haben wir Messdaten für Drehzahl, Spannung und Stromstärke aufgezeichnet und aus letzteren die Leistung berechnet. Die *Leistung* ist somit die *abhängige Größe*, die *Drehzahl* die *unabhängige*. Entsprechend würde ein Diagramm aus den beiden Datensätzen auf der x-Achse die Drehzahl und auf der y-Achse die Leistung zeigen.

Linien werden in solchen Diagramm überlicherweise für Funktionskurven eingesetzt, wie zum Beispiel einer Ausgleichskurve. Darauf wird später noch einmal eingegangen.

Ein gutes Diagramm sollte folgendes enthalten:

* angemessene Darstellung (Punkte, Linien, Balken etc.)
* relevanter Datenbereich wird dargestellt
* Achsenbeschriftungen mit Einheiten
* eindeutige Legende (vor Allem, wenn mehrere Datensätze in einem Diagramm dargestellt werden)



Die Darstellung von Graphiken und Diagrammen ist keine Standardfunktion von Python - genau wie zuvor das Einlesen von Daten. Python-Nutzer haben jedoch dafür Abhilfe geschaffen und Module entwickelt, die einfach angewendet werden können; genau wie es mit `numpy` im vorherigen Kapitel war.

Diese Variationsmöglichkeiten sind vor Allem dann sinnvoll, wenn mehrere Graphen in einem Diagramm angezeigt werden soll. Dies geschieht in Python, indem die `plt.plot()` Befehle in aufeinanderfolgende Zeilen geschrieben werden, ohne eine Leerzeile.

```python
plt.plot(Drehzahl,Leistung,"o")
plt.plot(Drehzahl,Leistung*2,"o", color="red")
plt.plot(Drehzahl,Leistung*3,"o", color="green")
```

Besonders in wissenschaftlichen Veröffentlichungen, aber auch in vielen Unternehmen wird mittlerweile großer Wert auf Barrierefreiheit gelegt. In unserem Fall bedeuted das, dass Farbenblinde oder Rot-Grün-Blinde nicht unterscheiden könnten welcher Datensatz zu welchem Diagramm gehört. Daher bietet es sich an, sowohl die Farbe, als auch die Form zu Variieren und Rot/Grüntöne zu meiden.

```python
plt.plot(Drehzahl,Leistung,"o")
plt.plot(Drehzahl,Leistung*2,"*", color="#00008B")
plt.plot(Drehzahl,Leistung*3,"p", color="#7D7D7D", alpha=0.5)
```

Wie Sie sehen, können nicht nur Farbnamen wie `red` oder `navy` verwendet, sondern auch spezifische Farbtöne durch ihren jeweiligen Hexcode verwendet werden (wie hier z.B.: `#7D7D7D` für einen Grauton steht). Auch können Farbpaletten (z.B. `C1` `C2`) oder ein RGB-Farbcode (`0.1, 0.2, 0.5`) verwendet werden. Wenn Sie eine bestimmte Farbe verwenden wollen, finden Sie im Internet die jeweiligen Bezeichnungen, ansonsten können Sie einfach die Farbnamen verwenden.

Außerdem können Sie durch das Argument `alpha` die Transparenz bestimmen mit einem Zahlenwert zwischen 0 und 1. Es gilt generell, dass in Python ein `,` Argumente und Funktionen trennt, Dezimalzahlen in Funktionen daher immer mit `.` Punkttrennung zu schreiben sind.

## Diagramm-Beschriftungen und Legenden erstellen

Ein Diagramm benötigt nicht nur Datenpunkte und zwei Achsen, sondern auch Achsenbeschriftungen und in manchen Fällen ein Diagrammtitel. Außerdem ist eine Legende nützlich, um Graphen voneinander unterscheiden zu können (siehe Beispiel oben). Diese können durch weitere `matplotlib`-Befehle erstellt werden.

> `plt.xlabel("")` erstellt eine x-Achsenbeschriftung
>
> `plt.ylabel("")` erstellt eine x-Achsenbeschriftung
> 
> `plt.title("")` erstellt einen Diagrammtitel

Für das obige Diagramm könnte das also wie folgt aussehen:

```python
plt.plot(Drehzahl,Leistung,"o")
plt.plot(Drehzahl,Leistung*2,"*", color="#00008B")
plt.plot(Drehzahl,Leistung*3,"p", color="#7D7D7D", alpha=0.5)
plt.xlabel("Drehzahl [min-1]")
plt.ylabel("Leistung [mW]")
plt.title("Drehzahl-Leistungsdiagramm dreier Windkrafträder")
```

Um eine Legende hinzuzufügen, wird die Funktion `plt.legend()` verwendet. Dafür muss die Bezeichnung des jeweiligen Graphen in der `plt.plot()`-Funktion über das Argument `label=""` übergeben werden. Die Legende wird automatisch so platziert, dass möglichst kein Datenpunkt verdeckt wird. Falls eine spezifische Orientierung der Legende gewünscht ist, kann im Argument von `plt.legend()` ein `loc=""` hinzugefügt werden. Die Ortsangabe kann dabei aus den Angaben `left right upper lower center` bestehen. Um eine Legende also oben rechts im Diagramm zu platzieren, lautet der Befehl `plt.legend(loc="upper right")`.

Nun fügen wir eine Legende dem oberen Diagramm hinzu.

```python
# Graphen
plt.plot(Drehzahl,Leistung,"o", label="WKA1")
plt.plot(Drehzahl,Leistung*2,"*", color="#00008B", label="WKA2")
plt.plot(Drehzahl,Leistung*3,"p", color="#7D7D7D", alpha=0.5, label="WKA3")

# Achsenbeschriftung
plt.xlabel("Drehzahl [min-1]")
plt.ylabel("Leistung [mW]")

# Diagrammbeschriftung
plt.title("Drehzahl-Leistungsdiagramm dreier Windkrafträder")

# Legende
plt.legend()
```

In dem Beispiel haben wir auf der x-Achse eine Drehzahl in der Einheit *Umdrehungen pro Minute*, die als min$^{-1}$ geschrieben wird, also mit einer Potenz. Um solche mathematischen Ausdrücke, die nicht auf der Standardtastatur vorhanden sind, auch in Beschriftungen in `matplotlib` muss die Formatierung nach der Mathematiksyntax in LaTeX erfolgen.

Um nun also die *-1* hochzustellen, nutzen wir `plt.xlabel(r"$ \mathregular{Drehzahl ~ [min^{-1}]}$")`.

Dieser Befehl im einzelnen bedeutet: `r` steht für *raw string*, die `$`-Zeichen Rahmen die Formel ein. `\mathregular{}` sorgt dafür, dass der Text nicht kursiv geschrieben wird, sondern Steil (Standardmäßig sind Formeln in LaTeX kursiv). Das `~`-Zeichen (Tilde) erzeugt ein Leerzeichen, da diese sonst in LaTeX-Formeln nicht vorkommen. Die Potenz wird schließlich durch `^{}` erzeugt. Generell sind in LaTeX die zugehörigen Daten in geschweiften Klammern. Fügen wir nun dies der Beschriftung hinzu erhalten wir das folgende Diagramm:

```python
# Graphen
plt.plot(Drehzahl,Leistung,"o", label="WKA1")
plt.plot(Drehzahl,Leistung*2,"*", color="#00008B", label="WKA2")
plt.plot(Drehzahl,Leistung*3,"p", color="#7D7D7D", alpha=0.5, label="WKA3")

# Achsenbeschriftung
plt.xlabel(r"$ \mathregular{Drehzahl ~ [min^{-1}]}$")
plt.ylabel("Leistung [mW]")

# Diagrammbeschriftung
plt.title("Drehzahl-Leistungsdiagramm dreier Windkrafträder")

# Legende
plt.legend()
```

## Achsenskalierung und Gitternetz

Die Achsenskalierung wird bei `matplotlib` automatisch so gewählt, dass alle Datenpunkte im Bereich liegen, mit einem kleinen Puffer zum Rand. Falls spezifische Achsenskalierungen gewünscht sind, können diese auch erstellt werden. Dafür werden die Befehle `plt.xlim( , )` und `plt.ylim( , )` verwendet wobei die linke und rechte Grenze (für `plt.xlim(linke Grenze, rechte Grenze)`) und die untere und obere Grenze (für `plt.ylim (untere Grenze, obere Grenze)`) angegeben werden können.

Durch den Befehl `plt.grid()` kann ein Gitternetz hinzugefügt werden.

```python
# Graphen
plt.plot(Drehzahl,Leistung,"o", label="WKA1")
plt.plot(Drehzahl,Leistung*2,"*", color="#00008B", label="WKA2")
plt.plot(Drehzahl,Leistung*3,"p", color="#7D7D7D", alpha=0.5, label="WKA3")

# Achsenbeschriftung
plt.xlabel(r"$ \mathregular{Drehzahl ~ [min^{-1}]}$")
plt.ylabel("Leistung [mW]")

# Diagrammbeschriftung
plt.title("Drehzahl-Leistungsdiagramm dreier Windkrafträder")

# Legende
plt.legend()

# manuelle Achsenskalierung
plt.xlim(2000, 5000)
plt.ylim(0, 500)

# Gitternetz
plt.grid()
```

Wenn nicht nur der Bereich der Achsen individualisiert werden soll, sondern auch die Unterteilung, so kann dies durch die Befehle `plt.xticks()` und `plt.yticks()` angepasst werden. Dafür können entweder manuell die Werte eingegeben werden, die auf der x- und y-Achse angezeigt werden sollen, oder es wird ein numpy-Array erstellt, dass genau diese Werte enthält.

ZSolche numpy-Arrays einfach zu erstellen gelingt mithilfe der Funkion `np.linspace(start, stop, Anzahl)`.

Die Syntax von `np.linspace()` sieht vor, dass ein Start und ein Endwert bestimmt werden und durch eine Anzahl an Schritten ergänzt wird. Die Funktion erzeugt dann automatisch Einträge in gleichmäßigen Schrittweiten, sodass genau die Anzahl an Werten vorhanden sind, die gefordert waren. Als Beispiel:

```python
np.linspace(2000, 5000, 7)
```

Durch die Argumente `(2000, 5000, 7)` werden 7 Einträge in einem Array erstellt, die zwischen 2000 und 5000 liegen.

Für unser Beispiel könnten wir nun das x- oder y-Raster unseren Wünschen entsprechend anpassen.

```python
# Graphen
plt.plot(Drehzahl,Leistung,"o", label="WKA1")
plt.plot(Drehzahl,Leistung*2,"*", color="#00008B", label="WKA2")
plt.plot(Drehzahl,Leistung*3,"p", color="#7D7D7D", alpha=0.5, label="WKA3")

# Achsenbeschriftung
plt.xlabel(r"$ \mathregular{Drehzahl ~ [min^{-1}]}$")
plt.ylabel("Leistung [mW]")

# Diagrammbeschriftung
plt.title("Drehzahl-Leistungsdiagramm dreier Windkrafträder")

# Legende
plt.legend()

# manuelle Achsenskalierung
plt.xlim(2000, 5000)
plt.ylim(0, 500)

# Gitternetz
plt.yticks(np.linspace(0, 500, 11))
plt.grid()
```

Außderdem können noch Zwischenstriche zu dem Diagramm hinzugefügt werden durch den Befehl `plt.minorticks_on()`.

```python
# Graphen
plt.plot(Drehzahl,Leistung,"o", label="WKA1")
plt.plot(Drehzahl,Leistung*2,"*", color="#00008B", label="WKA2")
plt.plot(Drehzahl,Leistung*3,"p", color="#7D7D7D", alpha=0.5, label="WKA3")

# Achsenbeschriftung
plt.xlabel(r"$ \mathregular{Drehzahl ~ [min^{-1}]}$")
plt.ylabel("Leistung [mW]")

# Diagrammbeschriftung
plt.title("Drehzahl-Leistungsdiagramm dreier Windkrafträder")

# Legende
plt.legend()

# manuelle Achsenskalierung
plt.xlim(2000, 5000)
plt.ylim(0, 500)

# Gitternetz
plt.yticks(np.linspace(0, 500, 11))
plt.minorticks_on()
plt.grid()
```

Die Größe des Diagramms kann durch den `plt.figure(figsize=(Breite, Höhe))`-Befehl auch selbst bestimmt werden. Die Werte für die Breite und Höhe sind dabei in inch anzugeben.

Außerdem können Sie dem Diagramm einen Text hinzufügen mit `plt.text(x-Koordinate, y-Koordinate, "Text")`.

```python
# Graphen
plt.plot(Drehzahl,Leistung,"o", label="WKA1")
plt.plot(Drehzahl,Leistung*2,"*", color="#00008B", label="WKA2")
plt.plot(Drehzahl,Leistung*3,"p", color="#7D7D7D", alpha=0.5, label="WKA3")

# Achsenbeschriftung
plt.xlabel(r"$ \mathregular{Drehzahl ~ [min^{-1}]}$")
plt.ylabel("Leistung [mW]")

# Diagrammbeschriftung
plt.title("Drehzahl-Leistungsdiagramm dreier Windkrafträder")

# Legende
plt.legend()

# manuelle Achsenskalierung
plt.xlim(2000, 5000)
plt.ylim(0, 500)

# Gitternetz
plt.yticks(np.linspace(0, 500, 11))
plt.minorticks_on()
plt.grid()

# Text im Diagramm
plt.text(3000, 475, "Punkt maximaler Leistung")
```

## Speichern von Diagrammen

Diagramme können durch den Befehl `plt.savefig("Dateiname.Endung")` direkt gespeichert werden. Als `Endung` können `.pdf`, `.png`, oder `.jpg` verwendet werden. Falls die ausgegebene Datei nicht so aussieht, wie erhofft, z.B. durch abgeschnittene Beschriftungen, kann der Befehl `plt.tight_layout()` verwendet werden.

```python
# Graphen
plt.plot(Drehzahl,Leistung,"o", label="WKA1")
plt.plot(Drehzahl,Leistung*2,"*", color="#00008B", label="WKA2")
plt.plot(Drehzahl,Leistung*3,"p", color="#7D7D7D", alpha=0.5, label="WKA3")

# Achsenbeschriftung
plt.xlabel(r"$ \mathregular{Drehzahl ~ [min^{-1}]}$")
plt.ylabel("Leistung [mW]")

# Diagrammbeschriftung
plt.title("Drehzahl-Leistungsdiagramm dreier Windkrafträder")

# Legende
plt.legend()

# manuelle Achsenskalierung
plt.xlim(2000, 5000)
plt.ylim(0, 500)

# Gitternetz
plt.yticks(np.linspace(0, 500, 11))
plt.minorticks_on()
plt.grid()

# Speichern des Diagramms als .png
plt.savefig("Data/Diagramm-1.png")
plt.tight_layout()
plt.savefig("Data/Diagramm-1_tight.png")
```

## Diagramme erstellen für Fortgeschrittene

Neben den oben genannten Personalisierungsmöglichkeiten gibt es weitere Funktionen, die hier kurz aufgegriffen werden.

### Schriftart und Größe

Schriftart und Größe können durch das Argument `fontsize=` und `family="Schriftfamilie"` bei jeder einzelnen Beschriftung eingestellt werden. Schriftfamilien sind z.B. `"serif"`, `"sans-serif"`, `"cursive"`, `"fantasy"`, und `"monospace"`.

Um eine bestimme Schriftart für alle Schriften zu verwenden, kann `plt.rc("font", family="Schriftfamilie")` und für die Schriftgröße `plt.rc("font", size=Schriftgröße)` verwendet werden.

### Hervorhebungen und Erkläuterungen

Vertikale Linien werden durch `plt.axvline(x=x-Wert)` und horizontale Linien durch `plt.axhline(y=y-Wert)` hinzugefügt.

Soll eine Fläche ins Diagramm eingezeichnet werden gelingt dies durch `plt.axvspan(xmin=minimaler x-Wert, xmax=maximaler x-Wert)` und `plt.axvhpan(ymin=minimaler y-Wert, ymax=maximaler y-Wert)` für horizontale und vertikale Bereiche, respektive. Diese können durch das Argument `label=""` auch in der Legende mit aufgenommen werden.`plt.axvspan(xmin=minimaler x-Wert, xmax=maximaler x-Wert)`.

Neben der oben dargestellten Art und Weise (funktionsorientierter Ansatz) Diagramme mit `matplotlib` zu erstellen gibt es auch eine andere Weise. Diese objektorientierte Darstellungsweise beitet sich vor Allem für sehr aufwändige Diagramme an. Eine Übersicht über die dazugehörigen Funktionen sehen Sie in folgenden Diagramm.

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.patches import Circle
from matplotlib.patheffects import withStroke
from matplotlib.ticker import AutoMinorLocator, MultipleLocator

royal_blue = [0, 20/256, 82/256]


# make the figure

np.random.seed(19680801)

X = np.linspace(0.5, 3.5, 100)
Y1 = 3+np.cos(X)
Y2 = 1+np.cos(1+X/0.75)/2
Y3 = np.random.uniform(Y1, Y2, len(X))

fig = plt.figure(figsize=(7.5, 7.5))
ax = fig.add_axes([0.2, 0.17, 0.68, 0.7], aspect=1)

ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.xaxis.set_minor_formatter("{x:.2f}")

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

ax.tick_params(which='major', width=1.0, length=10, labelsize=14)
ax.tick_params(which='minor', width=1.0, length=5, labelsize=10,
               labelcolor='0.25')

ax.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

ax.plot(X, Y1, c='C0', lw=2.5, label="Blue signal", zorder=10)
ax.plot(X, Y2, c='C1', lw=2.5, label="Orange signal")
ax.plot(X[::3], Y3[::3], linewidth=0, markersize=9,
        marker='s', markerfacecolor='none', markeredgecolor='C4',
        markeredgewidth=2.5)

ax.set_title("Anatomy of a figure", fontsize=20, verticalalignment='bottom')
ax.set_xlabel("x Axis label", fontsize=14)
ax.set_ylabel("y Axis label", fontsize=14)
ax.legend(loc="upper right", fontsize=14)


# Annotate the figure

def annotate(x, y, text, code):
    # Circle marker
    c = Circle((x, y), radius=0.15, clip_on=False, zorder=10, linewidth=2.5,
               edgecolor=royal_blue + [0.6], facecolor='none',
               path_effects=[withStroke(linewidth=7, foreground='white')])
    ax.add_artist(c)

    # use path_effects as a background for the texts
    # draw the path_effects and the colored text separately so that the
    # path_effects cannot clip other texts
    for path_effects in [[withStroke(linewidth=7, foreground='white')], []]:
        color = 'white' if path_effects else royal_blue
        ax.text(x, y-0.2, text, zorder=100,
                ha='center', va='top', weight='bold', color=color,
                style='italic', fontfamily='monospace',
                path_effects=path_effects)

        color = 'white' if path_effects else 'black'
        ax.text(x, y-0.33, code, zorder=100,
                ha='center', va='top', weight='normal', color=color,
                fontfamily='monospace', fontsize='medium',
                path_effects=path_effects)


annotate(3.5, -0.13, "Minor tick label", "ax.xaxis.set_minor_formatter")
annotate(-0.03, 1.0, "Major tick", "ax.yaxis.set_major_locator")
annotate(0.00, 3.75, "Minor tick", "ax.yaxis.set_minor_locator")
annotate(-0.15, 3.00, "Major tick label", "ax.yaxis.set_major_formatter")
annotate(1.68, -0.39, "xlabel", "ax.set_xlabel")
annotate(-0.38, 1.67, "ylabel", "ax.set_ylabel")
annotate(1.52, 4.15, "Title", "ax.set_title")
annotate(1.75, 2.80, "Line", "ax.plot")
annotate(2.25, 1.54, "Markers", "ax.scatter")
annotate(3.00, 3.00, "Grid", "ax.grid")
annotate(3.60, 3.58, "Legend", "ax.legend")
annotate(2.5, 0.55, "Axes", "fig.subplots")
annotate(4, 4.5, "Figure", "plt.figure")
annotate(0.65, 0.01, "x Axis", "ax.xaxis")
annotate(0, 0.36, "y Axis", "ax.yaxis")
annotate(4.0, 0.7, "Spine", "ax.spines")

# frame around figure
fig.patch.set(linewidth=4, edgecolor='0.5')
plt.show()
```
