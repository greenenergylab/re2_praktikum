### Code

Code kann ebenfalls in Markdown geschrieben werden, ohne dass dieser jedoch ausgeführt wird, indem der Code in Anführungszeichen ` `` ` eingerahmt wird.

Code kann aber auch mit verschiedenen Syntax-Highlights dargestellt werden. In diesem Praktikum insbesondere auch mit Python Highlighting, welches die Lesbarkeit erhöht. Die geschieht durch 3 Anführungszeichen aufeinanderfolgend mit dem Wort `python`  direkt folgend: ` ```python Code ``` `.

Beispiel:

```python
print("Guten Tag")
```

---

### Zeilenumbrüche

Text, der nur durch einen Zeilenumbruch im Code voneinander getrennt ist, 
zählt zu einem Paragraph.

Text, der durch 2 oder mehr Zeilenbrüche im Code voneinander getrennt ist,

zählt zu einem neuen Paragraph.

### $\LaTeX$

In Markdown können ebenfalls LaTeX-Umgebungen (Eigenschreibweise: $\LaTeX$) erzeugt und genutzt werden. Dabei wird alles was zwischen zwei Dollar-Zeichen (`$ $`) steht als LaTeX-Code interpretiert. $\LaTeX$ ist eine Sprache, die vor Allem für MINT-Fächer entwickelt wurde und die Darstellung von Formeln und Gleichungen in nahezu unbegrenzten Möglichkeiten ermöglicht. Für weitere Informationen zu $\LaTeX$-Syntax suchen Sie im Internet oder nutzen dieses [LaTeX-Cheatsheet](https://tug.ctan.org/info/undergradmath/undergradmath.pdf "LaTeX Cheatsheet").

ein Beispiel: Der folgende $\LaTeX$-Code:

```latex
$P_\text{el} = U * I $
```

führt zu Erzegung der entsprechenden Formel:

$P_\text{el} = U * I $

In der $\LaTeX$-Mathematikumgebung (durch `$ $` angezeigt) werden sämtliche Buchstaben kuriv geschrieben (als Variablen). Wenn Sie entsprechend Sub- oder Superskripte steil schreiben möchten, können Sie dies durch `$\text{steil geschriebener Text}$` tun.

Subskripte werden mit `_` begonnen, Superskripte mit `^`. Wenn Sie nur einen Buchstaben oder eine Zahl Hoch- oder Tiefstellen wollen, können Sie diesen direkt auf `_` oder `^` folgend schreiben. Wenn der Sub- oder Superskript, wie im Beispiel oben gezeigt, mehrere Buchstaben oder Zeichen enthalten soll, müssen Sie diesen mit geschweiften Klammern `{}`einrahmen. Ein Beispiel:

`$ P_\text{el}^3 $` führt zu:

$ P_\text{el}^3 $

und 

`$ P_\text{el}^{23} $` führt zu:

$ P_\text{el}^{23} $

wohingegen 

`$ P_el^23 $` zu:

$ P_el^23 $

führt.

> Weitere Informationen zur $\LaTeX$-Syntax finden Sie auch im Kapitel [Protokollerstellung](Protokollerstellung.ipynb).

### HTML

Markdown unterstützt ebenfalls HTML-Code, auch für die Formatierung. So führt zum Beispiel `<b>Fetter Text</b>` zu <b>Fetter Text</b> ebenso wie es `**Fetter Text**` zu **Fetter Text** tut. Innerhalb einer HTML-Umgebung funktioniert jedoch **keine** Markdown-Syntax.

So kann ***Fetter, kursiver Text*** erstellt werden durch: `***Fetter, kursiver Text***` (nur Markdown) oder durch `<b><i>Fetter, kursiver Text</i></b>` (nur HTML) oder durch `**<i>Fetter, kursiver Text</i>**` (HTML in Markdown), jedoch nicht durch `<b>*Fetter, kursiver Text*</b>` (Markdown in HTML).
